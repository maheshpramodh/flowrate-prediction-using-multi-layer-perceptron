```python
from sklearn.model_selection import train_test_split
import pandas as pd

x_train,x_test,y_train,y_test = train_test_split(df['ΔP'],df['qv'],test_size=.3,random_state=1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

LR = LinearRegression()
LR.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred = LR.predict(np.array(x_test).reshape(-1,1))
plt.scatter(x_train,y_train,c='m')
plt.plot(x_test,y_pred)
plt.xlabel('DIFFERENTIAL PRESSURE-(ΔP)')
plt.ylabel('FLOWRATE-(QV)')
y_pred
```

```python
from sklearn.metrics import mean_squared_error
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
from math import sqrt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np


LR = LinearRegression()
x_train = np.array(x_train).reshape(-1,1)
y_train =  np.array(y_train).reshape(-1,1)
LR.fit(x_train,y_train)
y_pred = LR.predict(np.array(x_test).reshape(-1,1))

SS = StandardScaler()
yt=SS.fit_transform(np.array(y_test).reshape(-1,1))
yp=SS.fit_transform(y_pred)
MSE = mean_squared_error(yt,yp)
RMSE = sqrt(mean_squared_error(yt,yp)) 
RSE = r2_score(yt,yp)
print("MSE: {}".format(MSE))
print("RMSE: {}".format(RMSE))
print("RSE: {}%".format(RSE*100))
```
